```python
# Visualiztion Script
import matplotlib.pyplot as plt
import numpy as np
import datetime

def histogram(data, x_label, title, save_name=None, show_count=True):
    """
    data: [(key, count)]
    """
    
    data = sorted(data, key=lambda x:x[1])  
    fig, ax = plt.subplots()
    ax.bar(range(len(data)), [t[1] for t in data]  , align="center")
    ax.set_xticks(range(len(data)))
    if show_count:
        ax.set_xticklabels(["{0} ({1})".format(t[0], t[1]) for t in data], rotation=45)
    else:
        ax.set_xticklabels([t[0] for t in data], rotation=45)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.grid(True)
    ax.set_ylabel("count")
    if save_name is not None:
        plt.savefig(save_name)
    else:
        plt.show()
    
def scatter_plot(data, x_label, title, reformat_label=False, line_format="r.-"):
    data = [x for x in data if x[0] is not None]
    data = sorted(data, key=lambda x:x[0])
    
    x = [d[0] for d in data]
    y = [d[1] for d in data]
    
    fig, ax = plt.subplots()
    ax.plot(x,y, line_format)
    ax.set_xlabel(x_label)
    ax.set_ylabel("count")
    ax.set_title(title)
    ax.grid(True)
    
    if reformat_label:
        labels = ax.get_xticks().tolist()
        labels = ["{0:.0f}".format(l) for l in labels]
        labels[0] = 'before 2004'
        ax.set_xticklabels(labels)

    plt.show()



histogram(boro_data, "BORO_NAME", "Distribution of Crimes Over Boro")
scatter_plot(year_data, "Year", "Distribution of Crimes Over Year", reformat_label=True)
scatter_plot(date_df, "Year", "Distribution of Crimes Over Time", line_format="r-")
    
```

```python
# Data Processing Script
def load_all_weather(data_dir="data/weather/"):
    """
    Function that loads all weather data
    """
    all_files = listdir(data_dir)
    df_list = []
    columns = ["date","f_high", "f_avg", "f_low", "dew_high", "dew_avg", "dew_low",
                         "h_high", "h_avg", "h_low","slp_high", "slp_avg", "slp_low",
                         "visbility_high", "visbility_avg", "visbility_low",
               "wind_high", "wind_avg", "wind_low" ,"precip", "events"]
    for data_file in all_files:
        if (data_file[0] != ".") and (len(data_file)==10):
            with open(os.path.join(data_dir, data_file), 'r') as f:
                data=f.read().replace('\r', '\n')
            with open(os.path.join(data_dir, data_file), 'w') as f:
                f.write(data)
            if data[3].isdigit():
                df = pd.read_csv(os.path.join(data_dir, data_file), sep=",", header=None, na_values=["-", "."])
            else:
                df = pd.read_csv(os.path.join(data_dir, data_file), sep=",", na_values=["-", "."])
            df.columns = columns
            df["year"]=int(data_file[:4])
            df["month"]=int(data_file[4:6])
            
            df["precip_is_T"]=df.apply(lambda x: x["precip"]=="T",1)
            df["precip"]=df.apply(lambda x: np.nan if x["precip"]=="T" else float(x["precip"]),1)
            df["is_rain"]=df.apply(lambda x: (not pd.isnull(x["events"])) and "rain" in x["events"].lower(),1)
            df["is_fog"]=df.apply(lambda x: (not pd.isnull(x["events"])) and "fog" in x["events"].lower(),1)
            df["is_snow"]=df.apply(lambda x: (not pd.isnull(x["events"])) and "snow" in x["events"].lower(),1)
            df["is_thunderstorm"]=df.apply(lambda x: (not pd.isnull(x["events"])) and "thunderstorm" in x["events"].lower(),1)
            df_list.append(df)
    return pd.concat(df_list)
            
def inspect_df(df):
    for column in df.columns:
        print(column)
        print(np.sum(pd.isnull(df[column]))/float(len(df)))
        print(np.unique(df[column]))
        print(len(np.unique(df[column])))
        print("\n\n")
    
```

```python
# Join functions

def find_close_point(geo_cord, kdtree, subway_locs):
    _, close_idx = kdtree.query(geo_cord, k=1)
    distance = vincenty(geo_cord, subway_locs[close_idx]).meters
    return close_idx, distance

def find_close_point_only(geo_cord, kdtree):
    _, close_idx = kdtree.query(geo_cord, k=1)
    return close_idx

def close_subway_lambda(x, kdtree, all_entrance_loc):
    close_idx, distance = find_close_point((float(x["Latitude"]), float(x["Longitude"])), kdtree, all_entrance_loc)
    return pd.Series({"idx":close_idx, "distance":distance})

def close_tract_lambda(x, kdtree):
    close_idx = find_close_point_only((float(x["Latitude"]), float(x["Longitude"])), kdtree)
    return pd.Series({"idx":close_idx})

def clean_dataset(df, columns, null_values):
    for col in columns:
        df = df[df.apply(lambda x: x[col] not in null_values,1)]
    df = df.dropna(subset=columns)
    return df

def create_clean_crime():
    crime_df = pd.read_csv("data/NYPD_Complaint_Data_Historic.csv")
    clean_crime = clean_dataset(crime_df, ["u'Latitude","Y_COORD_CD", "OFNS_DESC"], [" ", "  "])
    clean_crime_need = clean_crime[["u'Latitude", "Y_COORD_CD", "OFNS_DESC"]]
    clean_crime_need.to_csv("data/crime_need.csv", index=False)
```
